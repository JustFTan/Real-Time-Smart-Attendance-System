```python
import cv2
import os

video_dir = "C:/Users/justf/Desktop/Sem 6/KP/FaceRecog/Dataset_Vid"
output_dir = "C:/Users/justf/Desktop/Sem 6/KP/FaceRecog/Dataset_Faces"
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

for video_file in os.listdir(video_dir):
    if not video_file.endswith(".mp4"):
        continue

    label = os.path.splitext(video_file)[0]
    person_output = os.path.join(output_dir, label)
    os.makedirs(person_output, exist_ok=True)

    cap = cv2.VideoCapture(os.path.join(video_dir, video_file))
    frame_count = 0
    saved_count = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Setiap 5 frame (hemat data)
        if frame_count % 5 == 0:
            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            faces = face_cascade.detectMultiScale(gray, 1.3, 5)

            for (x, y, w, h) in faces:
                face_img = gray[y:y+h, x:x+w]
                face_img = cv2.resize(face_img, (100, 100))  # Ukuran seragam
                face_path = os.path.join(person_output, f"{saved_count}.jpg")
                cv2.imwrite(face_path, face_img)
                saved_count += 1
                break  # ambil satu wajah saja per frame

        frame_count += 1

    cap.release()
    print(f"{label}: {saved_count} wajah disimpan.")

print("Selesai ekstraksi semua video.")
```

```python
import cv2
import os
from PIL import Image

def video_to_dataset(video_path, output_dir, person_name="Unknown", skip_frames=5, max_faces=100):
    # Buat folder output untuk orang tersebut
    save_path = os.path.join(output_dir, person_name)
    os.makedirs(save_path, exist_ok=True)

    # Load video
    cap = cv2.VideoCapture(video_path)
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

    frame_count = 0
    saved_count = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret or saved_count >= max_faces:
            break

        if frame_count % skip_frames == 0:
            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)

            for (x, y, w, h) in faces:
                face = frame[y:y+h, x:x+w]
                face = cv2.resize(face, (160, 160))  # Ukuran ideal untuk MTCNN + FaceNet
                face_rgb = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
                img = Image.fromarray(face_rgb)

                filename = os.path.join(save_path, f"{saved_count}.jpg")
                img.save(filename)
                saved_count += 1

                print(f"✅ Saved: {filename}")
                break  # hanya simpan 1 wajah per frame (hindari double detect)

        frame_count += 1

    cap.release()
    print(f"🎉 Done! Total {saved_count} wajah disimpan di: {save_path}")
```
